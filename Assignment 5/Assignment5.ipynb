```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize
from nltk import pos_tag

# Download required resources (run once)
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')

# ---------------------------------
# Input text
# ---------------------------------
text = "The good teacher drives a car but avoids bad behavior."

tokens = word_tokenize(text.lower())
pos_tags = pos_tag(tokens)

# ---------------------------------
# POS mapping (Treebank → WordNet)
# ---------------------------------
def map_pos(tag):
    if tag.startswith('J'):
        return wn.ADJ
    elif tag.startswith('N'):
        return wn.NOUN
    else:
        return None   # Ignore verbs, adverbs, etc.

# ---------------------------------
# Semantic analysis function
# ---------------------------------
def semantic_analysis(word, pos):
    synsets = wn.synsets(word, pos=pos)
    if not synsets:
        return None

    synset = synsets[0]

    # Clean synonyms (limit to 3, remove word itself)
    synonyms = []
    for l in synset.lemmas():
        name = l.name().replace("_", " ")
        if name != word and name not in synonyms:
            synonyms.append(name)
        if len(synonyms) == 3:
            break

    # Antonyms
    antonyms = set()
    for l in synset.lemmas():
        for ant in l.antonyms():
            antonyms.add(ant.name().replace("_", " "))

    # Hypernyms
    hypernyms = [h.name().split(".")[0].replace("_", " ")
                 for h in synset.hypernyms()]

    return synonyms, antonyms, hypernyms


# ---------------------------------
# Analyze text
# ---------------------------------
seen = set()

for word, tag in pos_tags:
    wn_pos = map_pos(tag)

    if wn_pos and word.isalpha() and word not in seen:
        seen.add(word)

        result = semantic_analysis(word, wn_pos)
        if result:
            synonyms, antonyms, hypernyms = result

            print(f"\nWord: {word} ({wn_pos})")
            print("-" * 40)

            if synonyms:
                print("Synonyms :", synonyms)
            if antonyms:
                print("Antonyms :", antonyms)
            if hypernyms:
                print("Hypernyms:", hypernyms)
```
