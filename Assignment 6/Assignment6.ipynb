```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
def setup_translator(source_lang="en", target_lang="hi"):
    """
    Initializes the model and tokenizer for the specified language pair.
    For Marathi, use target_lang="mr"
    """
    model_name = f"Helsinki-NLP/opus-mt-{source_lang}-{target_lang}"
    
    print(f"Loading model: {model_name}...")
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)
    
    return tokenizer, model
```

```python
def translate_content(text, tokenizer, model):
    """
    Translates the input text using the loaded model.
    """
    # Tokenize the input text
    encoded_text = tokenizer(text, return_tensors="pt", padding=True)
    
    # Generate translation using the model
    translated_tokens = model.generate(**encoded_text)
    
    # Decode the tokens back to string
    translated_text = [tokenizer.decode(t, skip_special_tokens=True) for t in translated_tokens]
    
    return translated_text[0]
```

```python
if __name__ == "__main__":
    # 1. Setup the system
    tokenizer, model = setup_translator(source_lang="en", target_lang="hi")
    
    # 2. Sample public information content
    public_notices = [
        "Please wear a mask to prevent the spread of the virus.",
        "The deadline for filing your income tax return has been extended.",
        "Heavy rainfall is expected in the coastal regions tomorrow."
    ]
    
    # 3. Translate and display
    print("\n--- Translation Results ---")
    for notice in public_notices:
        hindi_translation = translate_content(notice, tokenizer, model)
        print(f"English: {notice}")
        print(f"Hindi:   {hindi_translation}\n")
```
