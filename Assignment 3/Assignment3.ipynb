```python
import nltk
import re
import math
import csv
from collections import Counter
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
documents = [
    "I love Natural Language Processing!",
    "NLP is very useful in real-world applications.",
    "Machine learning helps in language understanding."
]

labels = ["Positive", "Neutral", "Positive"]
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

cleaned_docs = [clean_text(doc) for doc in documents]
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

processed_docs = []
```

```python
for doc in cleaned_docs:
    tokens = word_tokenize(doc)
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    processed_docs.append(tokens)
```

```python
unique_labels = list(set(labels))
label_encoding = {label: idx for idx, label in enumerate(unique_labels)}
encoded_labels = [label_encoding[label] for label in labels]
```

```python
df = Counter()
for doc in processed_docs:
    df.update(set(doc))

N = len(processed_docs)
tfidf_docs = []

for doc in processed_docs:
    tf = Counter(doc)
    tfidf = {}
    for word in tf:
        tf_val = tf[word] / len(doc)
        idf_val = math.log(N / df[word])
        tfidf[word] = tf_val * idf_val
    tfidf_docs.append(tfidf)
```

```python
with open("cleaned_text.txt", "w") as f:
    for doc in cleaned_docs:
        f.write(doc + "\n")

# Save processed tokens
with open("processed_text.txt", "w") as f:
    for doc in processed_docs:
        f.write(" ".join(doc) + "\n")

# Save label encoding
with open("label_encoding.csv", "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["Label", "Encoded Value"])
    for label, value in label_encoding.items():
        writer.writerow([label, value])
```

```python
# Save TF-IDF values
with open("tfidf_output.txt", "w") as f:
    for i, tfidf in enumerate(tfidf_docs, 1):
        f.write(f"Document {i}: {tfidf}\n")

print("All preprocessing steps completed successfully.")
print("Outputs saved to files.")
```
