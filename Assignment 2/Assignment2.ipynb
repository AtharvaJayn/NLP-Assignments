```python
import nltk
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec

nltk.download('punkt')
```

```python
corpus = [
    "I love natural language processing",
    "Natural language processing is fun",
    "I love machine learning"
]
```

```python
print("Corpus:")
for i, doc in enumerate(corpus, 1):
    print(f"{i}. {doc}")

print("\n" + "-"*60)
```

```python
count_vectorizer = CountVectorizer()
bow_counts = count_vectorizer.fit_transform(corpus)
```

```python
print("Bag-of-Words (Count Occurrence):")
print(count_vectorizer.get_feature_names_out())
print(bow_counts.toarray())

print("\n" + "-"*60)
```

```python
bow_normalized = bow_counts.toarray().astype(float)
```

```python
for i in range(len(bow_normalized)):
    bow_normalized[i] /= bow_normalized[i].sum()
```

```python
print("Bag-of-Words (Normalized Count):")
print(bow_normalized)
print("\n" + "-"*60)
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)
```

```python
print("TF-IDF:")
print(tfidf_vectorizer.get_feature_names_out())
print(tfidf_matrix.toarray())
print("\n" + "-"*60)
```

```python
tokenized_corpus = [nltk.word_tokenize(doc.lower()) for doc in corpus]

word2vec_model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=50,
    window=3,
    min_count=1,
    workers=1
)
```

```python
print("Word2Vec Embedding for word 'language':")
print(word2vec_model.wv['language'])
```
