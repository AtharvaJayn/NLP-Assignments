```python
import nltk #Natural Language toolkit
from nltk.tokenize import (
    word_tokenize,
    WhitespaceTokenizer,
    TreebankWordTokenizer,
    TweetTokenizer,
    MWETokenizer
)
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
```

```python
text = "NLTK is amazing! I love Natural Language Processing. Let's learn NLP together :) #AI"
```

```python
print("Original Text:\n", text)
print("-" * 50)
```

```python
# 1. Whitespace Tokenization
wt = WhitespaceTokenizer()
print("Whitespace Tokenization:")
print(wt.tokenize(text))
print("-" * 50)
```

```python
# 2. Punctuation-based Tokenization
print("Punctuation-based Tokenization:")
print(word_tokenize(text))
print("-" * 50)
```

```python
# 3. Treebank Tokenization
treebank = TreebankWordTokenizer()
print("Treebank Tokenization:")
print(treebank.tokenize(text))
print("-" * 50)
```

```python
# 4. Tweet Tokenization
tweet_tokenizer = TweetTokenizer()
print("Tweet Tokenization:")
print(tweet_tokenizer.tokenize(text))
print("-" * 50)
```

```python
# 5. MWE (Multi-Word Expression) Tokenization
mwe = MWETokenizer([('Natural', 'Language', 'Processing')], separator='_')
print("MWE Tokenization:")
print(mwe.tokenize(word_tokenize(text)))
print("-" * 50)
```

```python
words = word_tokenize(text)
```

```python
# Porter Stemmer
porter = PorterStemmer()
print("Porter Stemmer:")
print([porter.stem(word) for word in words])
print("-" * 50)

```

```python
# Snowball Stemmer
snowball = SnowballStemmer("english")
print("Snowball Stemmer:")
print([snowball.stem(word) for word in words])
print("-" * 50)
```

```python
lemmatizer = WordNetLemmatizer()
print("Lemmatization:")
print([lemmatizer.lemmatize(word) for word in words])
print("-" * 50)
```
